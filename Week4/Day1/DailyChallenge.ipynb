# Global Weather Temperature Analysis

This project analyzes synthetic global weather data using NumPy, Pandas, and Matplotlib.  
The goal is to explore temperature trends across different cities and months and visualize the results.

```python
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Data Preparation

# Generating random temperatures

# low = -5
# high = 35
# size = (10, 12)

temperatures = np.random.uniform(-5, 35, (10, 12))

print(temperatures)
print(temperatures.shape)
```

```python
# Adding city names and months

cities = [
    'Helsinki', 'London', 'Tel Aviv', 'Baghdad', 'Tokyo',
    'New Delhi', 'Lagos', 'Cape Town', 'New York', 'Valparaiso'
]

months = [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
]
```

```python
# Converting to Pandas DataFrame

weather_df = pd.DataFrame(temperatures, index=cities, columns=months)

print(weather_df)
```

2. Data Analysis:

    Hint 1: Calculate the annual average temperature using DataFrame.mean(axis).

    Hint 2: Find the city with the highest and lowest average temperature using idxmax() and idxmin() methods.

    Calculate the annual average temperature for each city.
    Identify the city with the highest and lowest average temperature for the year.


```python
## Data Analysis

# Calculating average temperature of the year per city

city_avg = weather_df.mean(axis=1)

print(city_avg)
```

```python
# Searching for the city with highest average temperature

hottest_city = city_avg.idxmax()

print("City with highest average temperature:", hottest_city)
```

```python
# Searching for the city with lowest average temperature

coldest_city = city_avg.idxmin()

print("City with lowest average temperature:", coldest_city)
```

```python
## Data Visualization

## Monthly Temperature Trends by City

# The following line chart shows how temperatures change across the months for each city.

# Transposing the table and visualizing it as a line chart

plt.figure(figsize=(10,6))

weather_df.T.plot(marker="o")

plt.title("Monthly Temperature Trends for Each City")
plt.xlabel("Months")
plt.ylabel("Temperature (°C)")

plt.legend(title="Cities")
plt.grid(True)

plt.show()
```

```python
## Average Annual Temperature by City

# This bar chart compares the average yearly temperature across the selected cities.

city_avg.plot(kind='bar', figsize=(10,6), color='skyblue')

plt.title("Average Annual Temperature by City")
plt.xlabel("City")
plt.ylabel("Temperature (°C)")

plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

```python
# Average temperature per month

# Calculating averages

monthly_avg = weather_df.mean(axis=0)
print(monthly_avg)
```

```python
# Visualizing average temperature per month

## Average Monthly Temperature Trend

# This line plot shows the overall seasonal temperature pattern averaged across all cities.

monthly_avg.plot(marker='o', figsize=(10,6), color='red')

plt.title("Average Temperature Trend Across Months")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")

plt.grid(True)

plt.show()
```

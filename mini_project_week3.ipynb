QUESTION 5. The Pareto Principle, also known as the 80/20 rule, is a concept derived from the work of Italian economist Vilfredo Pareto. It states that roughly 80% of the effects come from 20% of the causes. For instance, identifying the top 20% of products that generate 80% of sales or the top 20% of customers that contribute to 80% of profit can help in prioritizing efforts and resources. This focus can lead to improved efficiency and effectiveness in business strategies. Can we apply Pareto principle to customers and Profit ? (Determine if 20% of the customers contribute to 80% of the profit.)

```python
# Grouping per customer
customer_profit = df.groupby("Customer Name")["Profit"].sum()
```

```python
# Sorting customers by profit
customer_profit = customer_profit.sort_values(ascending=False)
```

```python
# Calculating profit
cumulative_profit = customer_profit.cumsum()
```

```python
# Calculating total profit
total_profit = customer_profit.sum()
```

```python
# Converting to percents
cumulative_percentage = cumulative_profit / total_profit
```

```python
# Customers reaching to 80%
pareto_80 = cumulative_percentage[cumulative_percentage <= 0.8]

num_top_customers = len(pareto_80)
total_customers = len(customer_profit)

percentage_customers = num_top_customers / total_customers

print(f"Top {num_top_customers} customers out of {total_customers}")
print(f"This is {percentage_customers:.2%} of customers")
```

```python
# Visualizing
plt.figure(figsize=(8,5))

plt.plot(cumulative_percentage.values)

plt.axhline(0.8, color="red", linestyle="--", label="80% profit")
plt.axvline(num_top_customers, color="green", linestyle="--", label="Top customers")

plt.title("Pareto Analysis: Customers vs Profit")
plt.xlabel("Customers (sorted)")
plt.ylabel("Cumulative Profit (%)")

plt.legend()
plt.show()
```

The approximate result of 20% of customers → 80% of profit shows that Pareto principle holds also concerning this company's profits.

QUESTION 8.
Plot the Cumulative curve in Sales by Customers. Can we apply Pareto principle to customers and Sales ?

```python
# Total sales per customer
customer_sales = df.groupby("Customer Name")["Sales"].sum()
```

```python
# Sorting
customer_sales = customer_sales.sort_values(ascending=False)
```

```python
# Computing cumulative sales
cumulative_sales = customer_sales.cumsum()
```

```python
# Converting to percents
total_sales = customer_sales.sum()

cumulative_percentage = cumulative_sales / total_sales
```

```python
# Pareto curve
plt.figure(figsize=(8,5))

plt.plot(cumulative_percentage.values)

plt.axhline(0.8, color="red", linestyle="--", label="80% Sales")

plt.title("Pareto Curve: Customers vs Sales")
plt.xlabel("Customers (sorted)")
plt.ylabel("Cumulative Sales (%)")

plt.legend()
plt.show()
```

```python
# Checking the 20% rule
num_customers = len(customer_sales)

top_20_percent = int(0.2 * num_customers)

sales_top_20 = cumulative_percentage.iloc[top_20_percent]

print(f"Top 20% customers generate {sales_top_20:.2%} of total sales")
```

Pareto principle does not apply in conection with the sales of the company.
